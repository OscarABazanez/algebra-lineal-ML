# Una imagen es una matriz

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image # Ayuda a leer las imagenes
```

## Imagen

```python
plt.style.use('classic')

imagen = Image.open("./imagenes/imagen_ejemplo_frida_bredesen.jpg")

plt.imshow(imagen)
```

# Matriz de la imagen

```python
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

### Obteniendo un vector

```python
imagen_matriz = np.array(list(imagen_gr.getdata(band=0)), float)
imagen_matriz
```

## Convirtiendolo a una matriz

```python
imagen_matriz.shape = (imagen_gr.size[1], imagen_gr.size[0])
imagen_matriz
```

```python
imagen_matriz.shape
```

```python
plt.imshow(imagen_matriz, cmap='gray')
```

## Experimentando con la matriz
Modificar sus valores se afectara la imagen?
En machine learning muchas veces por optimizacion y para que nuestros algorimos no tengas problemas de convergencia. 

Nos conviene que todos los valores esten entre **0** y **1**. Para lograrlo en una imagen tendriamos que dividir todos los numeros entre los maximo posible.
- Maximo **255**: Nos permite normalizar los valores entre el 0 y el 1.
- Minimo **0**

```python
imagen_matriz_2 = imagen_matriz/255
imagen_matriz_2
```

```python
plt.imshow(imagen_matriz_2, cmap='gray')
```

```python
print(f'Maximo matriz 1: {np.max(imagen_matriz_2)}')
print(f'Maximo matriz 2: {np.max(imagen_matriz)}')
```

```python
print(f'Maximo matriz 1: {np.min(imagen_matriz_2)}')
print(f'Maximo matriz 2: {np.min(imagen_matriz)}')
```
